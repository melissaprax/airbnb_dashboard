```python
import sqlite3
from sqlite3 import Error
database_file_location = "./data/suicide.db"
```

```python
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
    except Error as e:
        print(e)

    return conn
```

```python
def select_all_suicides(conn, year):
    """
    Query all rows in the suicide table
    :param conn: the Connection object
    :return:
    """
    if year == '2019':
        cur = conn.cursor()
        cur.execute("SELECT country, s2019 FROM suicides")
    if year == '2020':
        cur = conn.cursor()
        cur.execute("SELECT country, s2020 FROM suicides")
        
    rows = cur.fetchall()

    for row in rows:
        print(row)
```

```python
def select_suicides_by_year(conn, country):
    """
    Query suicides by country
    :param conn: the Connection object
    :param country:
    :return:
    """
    cur = conn.cursor()
    cur.execute("SELECT * FROM tasks WHERE country=?", (country,))

    rows = cur.fetchall()

    for row in rows:
        print(row)
```

```python
def main():
    database = database_file_location
    year = '2020'
    # create a database connection
    conn = create_connection(database)
    with conn:
        print(f"2. Query suicides by {year}")
        select_all_suicides(conn, year)
```

```python
if __name__ == '__main__':
    main()
```
